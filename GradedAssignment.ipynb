#Tesla Dashboard

```python
tesla_data = yf.download('TSLA', start='2020-01-01', end='2024-01-01')
```

```python
tesla_data.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 4))
plt.plot(tesla_data.index, tesla_data['Close'], label='Tesla Stock Closing Price', color='blue')
plt.title('Tesla Stock Closing Price')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```
